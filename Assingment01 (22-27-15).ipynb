```python
# question 05 | code 
# NAME : MOHIT KUMAR | REG.NO.: 22-27-15| BRANCH : DATA SCIENCE 
num=int(input("enter the no of element to find  factorial "))
def factorial(n):
  if (n==1 or n==0):
    return 1
  else:
    return (n*factorial(n-1))

result= factorial(num)
print(f"factorial {num}! =" , result)
```
